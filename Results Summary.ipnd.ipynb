# Student Performance Prediction Report

## 1. Goal
The goal was to understand which features influence performance and to build a machine learning model that can help identify the final grade of the students and ackowledge who may need extra support.

**Target Variable:** 
Final grade

## 2. Model Used in this prediction
The machine learning classification model that was trained to predict the final grade was Random Forest.

## 3. Model Evaluation Results
The model was evaluated using accuracy, precision, recall, and F1 score.
The final results ganered were:

- **Accuracy:** 0.38  
- **Precision:** 0.41  
- **Recall:** 0.38  
- **F1 Score:** 0.37  

## 4. Interpretation of Results
The results show that the model struggled to identify strong patterns in the dataset.  
The low accuracy and F1 score suggest that:
- Many predictions were incorrect  
- More feature engineering or a different model may be needed  
Although the model performance was low, the project demonstrates how data preparation, feature engineering, and evaluation work in practice.
